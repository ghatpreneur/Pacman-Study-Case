# Data Collection and data Processing

```python
# loading dataset
path = 'Copy of sonar data.csv'

# header None karena didatanya emang tidak ada headernya
sonar_data = pd.read_csv(path, header=None)
sonar_data
```

# Memisahkan data dan Label

```python
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]
```

```python
print(X)
```

```python
print(Y)
```

Training and Test data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
'''fungsi stratify adalah untuk menjamin distribusi kelas yang sama antara set pelatihan (training set) dan set pengujian (testing set).
stratify digunakan pada saat atau pada kasus data yang tidak seimbang (imbalanced data) dimana salah satu kelas memiliki jumlah yang jauh lebih banyak daripada kelas lainya'''
```

```python
# melihat dimensi data pelatihan 
print(X.shape, X_train.shape,X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

Model Training --> LOREG

```python
model = LogisticRegression()
```

```python
#training logistic regression model with training data
model.fit(X_train, Y_train)
```

Model Evaluation

```python
# Accuracy on the training data
X_train_predicion = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_predicion, Y_train)
```

```python
print(f"Accuracy untuk mempredeksi mine vs rock sebesar : {training_data_accuracy}")
```

```python
# Accuracy on test data
X_test_predicion = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_predicion, Y_test)
```

```python
print(f"Accuracy untuk mempredeksi mine vs rock sebesar : {test_data_accuracy}")
```
