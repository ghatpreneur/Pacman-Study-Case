```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import  seaborn as sns
```

# Data Load

```python
google = pd.read_csv('googleplaystore.csv')
```

```python
google.head()
```

```python
google.info()
```

```python
# mengatasi missing values
google.dropna(inplace=True) # menghapus colums missing values

# Remove Duplicates column
google.drop_duplicates(inplace=True)
```

```python
google.info()
```

```python
google.duplicated().sum()
```

menangani data yang tidak sesuai representasi

```python
google['Installs'] = google['Installs'].str.replace('+', '') # menghapus '+'google.ipynb
google['Installs'] = google['Installs'].str.replace(',', '') # menghapus '+'google.ipynb

# mengubah ke int
google['Installs'] = google['Installs'].astype('Int64') 

# ubah data type reviews ke numerical columns
google['Reviews'] = google['Reviews'].astype('Int64')
```

```python
google.head()
```

# Data Visualization

tujuanya adalah:
- melihat distribusi data
- melihat hubungan antar data


```python
# Barplot visualization
plt.figure(figsize=(10, 8))
sns.countplot(data=google, x='Category')
plt.xlabel('Category')
plt.xticks(rotation = 90)
plt.ylabel("Count")
plt.title('Distribusi kategori app')
plt.show()
```

```python
# Scatter plot hubungan antara Reviews dan Rating
plt.figure(figsize=(10, 8))
sns.scatterplot(data=google, x='Reviews', y='Rating')
plt.xlabel('Number Of Reviews')
plt.ylabel('Rating')
plt.title("Korelasi Antara reviews dan Rating") 
plt.show()
```

```python
# Boxplot
plt.figure(figsize=(10, 8))
sns.boxplot(data=google, x='Category', y='Installs',)
plt.xlabel('Category')
plt.ylabel('Installs')
plt.xticks(rotation=90)
plt.title('Distribusi antara Category dan Install')
plt.show()
```

# Hypotesis and Statistical

```python
google.head()
```

```python
from scipy.stats import ttest_ind

# scaling the dataset into free and paid apps
free_apps = google[google['Type'] == 'Free']
paid_apps = google[google['Type'] == 'Paid']
```

```python
# performing t-test
t_stat, p_value, = ttest_ind(free_apps['Rating'], paid_apps['Rating']) 
```

```python
p_value
```

```python
# Checking the P value
if p_value < 0.05:
 print("Reject null hypotesis: There is a significant difference in ratings.")
else:
 print("Fail to reject null hypotesis: There is no significant difference in ratings")
```

# Data Storytellig 

```python
# Scatter plot of Ratings vs Installs
plt.figure(figsize=(10, 8))
sns.scatterplot(data=google, x='Rating', y='Installs', hue='Category')
plt.xlabel('Rating')
plt.ylabel("Number Of Installs")
plt.title('Relationship between Ratings and Installs')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
# Line plot Of Average Rating Over Time
plt.figure(figsize=(10, 8))
sns.lineplot(data=google, x='Last Updated', y='Rating')
plt.xlabel('Last Update')
plt.xticks(rotation=90)
plt.ylabel('Average Rating')
plt.title('Average Rating Trend Over Time')
plt.show()
```

```python
google.head()
```

```python
# Convert 'Last Updated' to datetime
google['Last Updated'] = pd.to_datetime(google['Last Updated'])

# Set 'Last Updated' as the index
google.set_index('Last Updated', inplace=True)

# Resample 'Rating' to monthly frequency
google_to_monthly = google['Rating'].resample('M').mean()
```

```python
# Line plot Of Average Rating Over Time
plt.figure(figsize=(10, 8))
sns.lineplot(data=google_to_monthly)
plt.xlabel('Last Update (Monthly)')
plt.xticks(rotation=90)
plt.ylabel('Average Rating')
plt.title('Average Rating Trend Over Time')
plt.show()
```
