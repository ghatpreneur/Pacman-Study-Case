# Analisis

```python
ai['Estimated_Valuation_Billion_USD'].value_counts(ascending=True)
```

distribusi kolom kategorikal

```python
kategorik_columns = ai.select_dtypes(exclude=['number']).columns


for col in  kategorik_columns:
 plt.figure(figsize=(10, 8))
 
 sns.countplot(data=ai, x=col)
 plt.title(f'Distribusi Frekuensi Untuk columns {col}', fontsize=12)
 plt.xticks(rotation=45, ha='right')
 plt.tight_layout()
 plt.show()
```

diatribusi kolom numerikal

```python
numerik_columns = ai.select_dtypes(include=['number']).columns

for col in  numerik_columns:
 plt.figure(figsize=(10, 8))
 
 sns.histplot(data=ai, x=col)
 plt.title(f'Distribusi Frekuensi Untuk columns {col}', fontsize=12)
 plt.xticks(rotation=45, ha='right')
 plt.tight_layout()
 plt.show()
```

```python
ai.columns
```

```python
ai_sorted = ai.sort_values(by='Estimated_Valuation_Billion_USD', ascending=False)
```

```python
top_10 = ai_sorted.head(10)
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(data=top_10, x='AI Name', y='Estimated_Valuation_Billion_USD')
plt.title('TOP 10 AI dengan Keuntungan Terbesar')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
ai.groupby('AI Name')['Estimated_Valuation_Billion_USD'].mean().sort_values(ascending=True)
```

```python
plt.figure(figsize=(10, 8))
sns.barplot(data=ai, x='AI Type', y='Estimated_Valuation_Billion_USD', color='purple')
plt.title("AI Type Yang paling menguntungkan")
plt.xticks(rotation=45, ha='right')
plt.show()
```
