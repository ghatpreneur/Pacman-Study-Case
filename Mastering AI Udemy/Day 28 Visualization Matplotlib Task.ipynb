# Matplotlib Design

| **Kategori**         | **Parameter/Kode**            | **Penjelasan Fungsi**                                                                 |
|-----------------------|-------------------------------|-------------------------------------------------------------------------------------|
| **Aesthetics**       | `color='#F5F547'`             | Mengatur warna elemen plot (misalnya garis, marker). Warna dapat ditentukan dalam format hex code. |
|                       | `linestyle='--'`              | Menentukan gaya garis pada plot (misalnya garis putus-putus `--`, garis solid `-`, dll.). |
|                       | `marker='o'`                  | Menambahkan marker pada titik data (misalnya lingkaran `'o'`, segitiga `'^'`, dll.). |
|                       | `alpha=0.7`                   | Mengatur tingkat transparansi elemen plot. Nilai `alpha` berkisar antara 0 (sepenuhnya transparan) hingga 1 (sepenuhnya buram). |
|                       | `figsize=(10, 6)`             | Menentukan ukuran gambar plot dalam satuan inci (lebar x tinggi). Misalnya `(10, 6)` berarti lebar 10 inci dan tinggi 6 inci. |
| **Annotations**      | `plt.title()`                 | Menambahkan judul untuk keseluruhan plot. Judul membantu menjelaskan konteks visualisasi. |
|                       | `plt.xlabel(), plt.ylabel()`  | Menambahkan label pada sumbu x dan y. Label ini mendeskripsikan arti dari data pada masing-masing sumbu. |
|                       | `plt.legend()`                | Menampilkan legenda untuk membedakan elemen-elemen plot yang berbeda (misalnya beberapa garis atau dataset). |
|                       | `plt.text(x, y, "text")`      | Menambahkan teks statis pada koordinat tertentu `(x, y)` di dalam plot. |
|                       | `plt.annotate()`              | Menambahkan anotasi dengan panah dan teks untuk menyoroti elemen tertentu dalam plot. |
| **Export & Polish**  | `plt.savefig("plot.png")`     | Menyimpan plot ke file dengan format tertentu (misalnya PNG, PDF, SVG). |
|                       | `dpi=300`                     | Mengatur resolusi gambar dalam dots per inch (DPI). Semakin tinggi DPI, semakin tajam gambar. |
|                       | `plt.subplot()`               | Membagi area plot menjadi beberapa subplot (grid) untuk menampilkan beberapa plot dalam satu gambar. |
|                       | `plt.grid(True)`              | Menambahkan grid pada plot untuk memudahkan pembacaan nilai pada sumbu x dan y. |
|                       | `plt.tight_layout()`          | Mengoptimalkan tata letak plot agar elemen-elemen seperti judul, label, dan legenda tidak saling tumpang tindih. |

1. Lineplot Harga Saham

```python
import pandas as pd
stock_df = pd.DataFrame({
    'Day': range(1, 11),
    'Stock_Price': [100, 105, 103, 108, 110, 107, 112, 115, 113, 118]
})
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))
plt.plot(stock_df['Day'], stock_df['Stock_Price'], color='green')
plt.title("LinePlot Harga Saham")
plt.xlabel('Day')
plt.ylabel('Stock Pirce')
plt.show()
```

2. Perbandingan Produk

```python
product_df = pd.DataFrame({
    'Product': ['A', 'B', 'C', 'D'],
    'Sales': [120, 95, 150, 80]
})
```

```python
plt.figure(figsize=(10, 8))
plt.bar(product_df['Product'], product_df['Sales'], color=['yellow', 'orange', 'red'])
plt.title('Membandingkan Penjualan Harga Produk')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.show()

# jika ingin plot horizonatal
# plt.barh
```

3. Analisis dan Distribusi

```python
import numpy as np
random_data = np.random.normal(50, 15, 500)  # mean=50, std=15
dist_df = pd.DataFrame({'Random_Number': random_data})
```

```python
plt.figure(figsize=(10, 8))
plt.hist(random_data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribusi Data')
plt.grid(True,alpha=0.3)
plt.show()
```

4. korelasi tinggi badan vs berat badan

```python
correlation_df = pd.DataFrame({
    'Height': [160, 165, 170, 175, 180, 155, 172, 168, 185, 158, 178, 162, 176, 169, 182],
    'Weight': [55, 60, 65, 70, 75, 50, 68, 62, 80, 53, 72, 57, 71, 64, 78]
})
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(correlation_df['Height'], correlation_df['Weight'], color='#F5F547', edgecolors='#242424', s=100, alpha=0.7)
plt.grid(True, alpha=0.3)
plt.title('Korelasi Antar columns numerik')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()

# jika ingin plot horizonatal
# plt.barh
```
